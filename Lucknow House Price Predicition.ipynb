```python
mean_price_per_location = df.groupby('location')['price_lakh'].mean().reset_index()
mean_price_per_location.columns = ['location', 'avg_price']

# Define the price bins
bins = [0, 10, 20, 30, 40, 50, 60, np.inf]  # Adjust these ranges as needed
labels = ['<10L', '10-20L', '20-30L', '30-40L', '40-50L', '50-60L', '>60L']

# Create the price range column
mean_price_per_location['price_range'] = pd.cut(mean_price_per_location['avg_price'], bins=bins, labels=labels)

# Merge the price range back to the original dataframe
df = df.merge(mean_price_per_location[['location', 'price_range']], on='location', how='left')

# Now, you can drop the 'location' column if you no longer need it
df.drop(columns=['location'], inplace=True)

# Optionally, you can convert 'price_range' to a numerical encoding if needed
df['price_range_encoded_acc_to_location'] = df['price_range'].cat.codes

print(df.head())
```
